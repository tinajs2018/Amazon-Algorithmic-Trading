```python
daily_close=df[['Adj Close']]
daily_close.head()
```

```python
daily_returns=daily_close.pct_change()
daily_returns.head()
```

```python
#replacing null values
daily_returns.fillna(0,inplace=True)
daily_returns.head()
```

```python
plt.figure(figsize=(15,5))
plt.plot(daily_returns)
plt.title('AMAZON daily returns from 2020 to 2021 ')
plt.xlabel('Dates')
plt.ylabel('Daily returns of AMAZON stock')
plt.grid()
plt.show()
```
